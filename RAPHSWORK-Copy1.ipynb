# 8.	Transform all features with categorical values to numerical values.

```python
titanic_disaster['sex_num'] = pd.factorize(titanic_disaster.Sex)[0]
titanic_disaster['cabin_num']= pd.factorize(titanic_disaster.Cabin)[0]
titanic_disaster['pclass_num'] = pd.factorize(titanic_disaster.Pclass)[0]
titanic_disaster['embarked_num'] = pd.factorize(titanic_disaster.Embarked)[0]
titanic_disaster['name_num'] = pd.factorize(titanic_disaster.Name)[0]
titanic_disaster['ticket_num'] = pd.factorize(titanic_disaster.Ticket)[0]


titanic_disaster
```
